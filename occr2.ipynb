```python
# Import necessary libraries
import cv2
import pytesseract
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import transforms, models
from google.colab import files  # For file upload and download
from docx import Document
from fpdf import FPDF
from PIL import Image
```

```python
def preprocess_image(image_path):
    image = cv2.imread(image_path)
    if image is None:
        return None

    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Increase contrast using CLAHE
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
    enhanced = clahe.apply(gray)

    # Apply bilateral filter to reduce noise
    filtered = cv2.bilateralFilter(enhanced, 9, 75, 75)

    # Apply adaptive thresholding to enhance text visibility
    thresh = cv2.adaptiveThreshold(filtered, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

    return thresh
```

```python
def extract_text(image_path):
    processed_image = preprocess_image(image_path)
    if processed_image is None:
        return "Error: Could not process image."

    # Use PSM (Page Segmentation Mode) 3 for better block detection
    extracted_text = pytesseract.image_to_string(processed_image, config="--psm 3")

    return extracted_text
```

```python
def upload_and_extract_text():
    uploaded = files.upload()  # Upload an image file
    image_path = list(uploaded.keys())[0]  # Get uploaded file name

    print("\nðŸ”¹ Extracting text from:", image_path)
    text = extract_text(image_path)  # Process image & extract text

    print("\nðŸ“œ **Extracted Text:**\n")
    print(text)

    return text
```

```python
def save_text(text, file_format="txt"):
    filename = "extracted_text." + file_format

    if file_format == "txt":
        with open(filename, "w", encoding="utf-8") as file:
            file.write(text)

    elif file_format == "docx":
        doc = Document()
        doc.add_paragraph(text)
        doc.save(filename)

    elif file_format == "pdf":
        pdf = FPDF()
        pdf.add_page()
        pdf.set_auto_page_break(auto=True, margin=15)
        pdf.set_font("Arial", size=14)
        pdf.multi_cell(0, 10, text)
        pdf.output(filename)

    print(f"\nâœ… **Saved as {filename}**")

    # Download the file
    files.download(filename)
```

```python
text_result = upload_and_extract_text()  # Upload & extract text
save_text(text_result, "txt")  # Save & Download as TXT
```
